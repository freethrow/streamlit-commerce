```python
import pandas as pd, numpy as np
```

```python
data=pd.read_excel('countriesMarch2021.xls', sheet_name='Tabela - 2021-04-28T082613.292', skiprows=2, usecols="B:F")
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
data['Indeks - izvoz']
```

```python
data.replace('*',0, inplace=True)
data.replace('-',0, inplace=True)
```

```python
pd.to_numeric(data['Indeks - izvoz'])
```

```python
data['exp_var'] = pd.to_numeric(data['Indeks - izvoz'])
data['imp_var'] = pd.to_numeric(data['Indeks - uvoz'])
```

```python
data.head()
```

```python
data['exp'] = pd.to_numeric(data['Izvoz'], downcast='float')
data['imp'] = pd.to_numeric(data['Uvoz'], downcast='float')
```

```python
data = data[['Naziv','exp','imp','exp_var','imp_var']]
```
